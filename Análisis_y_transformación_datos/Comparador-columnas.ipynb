```python
import pandas as pd
```

```python
df= pd.read_csv('raw_data.csv')

def rawdata(df, columna):
    valores_unicos = df[columna].nunique()
    num_registros = len(df[columna])
    duplicados = num_registros - valores_unicos

    print(f"Valores únicos: {valores_unicos}\nNúmero de registros: {num_registros}\nRegistros duplicados: {duplicados}")
    print('----------------')
    print("\nPorcentajes:")
    print(df[columna]
        .value_counts(normalize=True, dropna=False)
        .mul(100)
        .round(2)
        .sort_values(ascending=False))
```

```python
# Con esta función hacemos la misma petición pero para el archivo transformado

dftrans= pd.read_csv('raw_data_limpio_2.csv')

def cambios(dftrans, columna):
    valores_unicos = dftrans[columna].nunique()
    num_registros = len(dftrans[columna])
    duplicados = num_registros - valores_unicos

    print(f"Valores únicos: {valores_unicos}\nNúmero de registros: {num_registros}\nRegistros duplicados: {duplicados}")
    print('----------------')
    print("\nPorcentajes:")
    print(df[columna]
        .value_counts(normalize=True, dropna=False)
        .mul(100)
        .round(2)
        .sort_values(ascending=False))
```

```python
# USO:

rawdata(df, "hourlyrate")
print('\n-----<------<-------<---@\n')
cambios(dftrans, "hourlyrate")
```
