```python
import pandas as pd
```

```python
df= pd.read_csv('raw_data.csv')
dftrans = pd.read_csv('raw_data_limpio_2.csv')

def rawdata(df, columna):
    valores_unicos = df[columna].nunique()
    num_registros = len(df[columna])
    duplicados = num_registros - valores_unicos

    print(f"Valores √∫nicos en archivo original: {valores_unicos}\nN√∫mero de registros: {num_registros}\nRegistros duplicados: {duplicados}")
    print('----------------')
    print("\nPorcentajes:")
    print(df[columna]
        .value_counts(normalize=True, dropna=False)
        .mul(100)
        .round(2)
        .sort_values(ascending=False))

```

```python
# Con esta funci√≥n hacemos la misma petici√≥n pero para el archivo transformado

dftrans= pd.read_csv('raw_data_limpio_2.csv')

def cambios(dftrans, columna):
        valores_unicos = dftrans[columna].nunique()
        num_registros = len(dftrans[columna])
        duplicados = num_registros - valores_unicos

        print(f"Valores √∫nicos en archivo transformado: {valores_unicos}\nN√∫mero de registros:{num_registros}\nRegistros duplicados: {duplicados}")

        print('----------------')

        print("\nPorcentajes:")
        print(dftrans[columna]
        .value_counts(normalize=True, dropna=False)
        .mul(100)
        .round(2)
        .sort_values(ascending=False))
```

```python
# USO:

rawdata(df, "worklifebalance")
print('\n-----<------<-------<---@\n')
cambios(dftrans, "worklifebalance")
# Se anade calculo de media, mediana, mode - util para imputaciones.
stats = dftrans["worklifebalance"].agg({
    'Mean': 'mean',
    'Median': 'median',
    'Mode': lambda x: x.mode().iloc[0]
})

print("üìä Estad√≠sticas de WorkLifeBalance")
print("---------------------------------")
print(f"Media : {stats['Mean']:.2f}")
print(f"Mediana : {stats['Median']:.2f}")
print(f"Moda : {stats['Mode']}")
```
