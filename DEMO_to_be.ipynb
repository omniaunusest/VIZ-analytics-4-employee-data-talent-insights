# Ideas DEMO:

### 1. Perspectiva Humanista:

Impacto de los datos en la vida de lxs empleadxs. La historia no es sobre optimización de costes, sino sobre la mejora de la experiencia laboral y la ética de la IA.

Somos: Equipo que está asegurando que la IA y el análisis se usen para humanizar la gestión del talento, no solo para automatizarla.

Clientx: Persona encargada de RRHH y el Departamento de Ética y Cultura de ABC Corporation.

Problema: "¿Estamos fallando en ver las señales de infelicidad?, ¿Nuestra tecnología (IA) está creando un ambiente de trabajo justo y equitativo?"

Narrativa de la Presentación: La demo es una Auditoría de Bienestar.

(¡Esta opción encajaría con la idea que planteó Patricia acerca de transformar la recogida de datos sobre ‘géneros’, para ser más inclusivas y éticas!) ;)

### 2. Perspectiva de laboratorio de I+D:

Historia no es sobre dar consejos, sino sobre probar que la tecnología del sistema de gestión de talento funciona.

Somos: Equipo de Investigación y Desarrollo (I+D) que trabaja internamente para optimizar un producto de ABC Corporation: el sistema de gestión de talento.

Clientx: Product Owner de la plataforma de selección inteligente

Problema: "¿Nuestros datos de empleadxs son lo suficientemente limpios y precisos como para entrenar futuros modelos de retención? ¿Podemos fiarnos de la información que la plataforma recopila?"

Narrativa: La demo es una sesión de control de calidad de los datos.

### 3. Perspectiva del Inversor:

Historia sobre optimización de costes, impacto financiero de la retención.

Somos: Equipo de Análisis de Rentabilidad que cuantifica el riesgo de pérdida de talento.

Clientx: Directora Financiera (CFO) o la Junta Directiva de ABC Corporation.

Problema: "¿Cuánto nos cuesta realmente la rotación y dónde debemos invertir cada euro para maximizar la retención y el crecimiento?"

Narrativa: La demo es una Evaluación de Riesgo Financiero.
